```python
from constrained_power_law_surrogates_module import gen_power_law_surr_list
import random
random.seed(0)
```

```python
#Basic uses of surrogate generation function:
seq = [9, 9, 3]

print('Default:')
surr_list = gen_power_law_surr_list(seq)
print(surr_list)

print('Constrained:')
surr_list = gen_power_law_surr_list(seq, surr_method='cons', N=2*len(seq), resamp_method='subs', x_min=3, num_surr=4)
print(surr_list)

print('Typical:')
surr_list = gen_power_law_surr_list(seq, surr_method='typi', N=2, resamp_method='boot', x_min=3, num_surr=9)
print(surr_list)

print('Known scale exponent:')
surr_list = gen_power_law_surr_list(seq, surr_method='know', x_min=5, num_surr=9, scale_exp=2)
print(surr_list)

print('Subsampled:')
surr_list = gen_power_law_surr_list(seq, surr_method='subs', x_min=2, num_surr=9)
print(surr_list)

print('Bootstrapped:')
surr_list = gen_power_law_surr_list(seq, surr_method='boot', x_min=2, num_surr=9)
print(surr_list)
```

```python
#Check that all options can be used without error:
seq = [9, 9, 3]
for surr_method in ['cons', 'typi', 'know', 'subs', 'boot']:
    print('Surrogate method: ' + '\'' + surr_method + '\'')
    for N in [0, 1, 2, 3, 10]:
        print('Final N: ' + str(N))
        for resamp_method in ['subs', 'boot']:
            print('Resampling: ' + '\'' + resamp_method + '\'')
            for x_min in [1, 2]:
                print('x_Min: ' + str(x_min))
                for num_surr in [0, 1, 2, 10]:
                    print('Number of surrogates: ' + str(num_surr))
                    for scale_exp in [1.5, 2.5, 3.5]:
                        print('Scale exponent (only used when surr_method==\'know\'): ' + str(scale_exp))
                        surr_list = gen_power_law_surr_list(seq, surr_method=surr_method, N=N, resamp_method=resamp_method, x_min=x_min, num_surr=num_surr, scale_exp=scale_exp)
                        print(surr_list)
```
